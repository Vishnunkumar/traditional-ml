```python
from statsmodels.tsa.api import SARIMAX, ExponentialSmoothing
from statsmodels.tsa import seasonal
import pandas as pd
import numpy as np
```

```python
train_df = pd.read_csv('Train.csv')
train_df['Datetime'] = pd.to_datetime(train_df['Datetime'])
test_df = pd.read_csv('Test.csv')
test_df['Datetime'] = pd.to_datetime(test_df['Datetime'])

train_df.index = train_df['Datetime']
test_df.index = test_df['Datetime']
test_df['Count'] = 0
```

```python
train_df = train_df.iloc[-2400:,:]
```

```python
p = 12
for i in range(0, int(len(test_df)/p)):
    print(i)
    exp_model = ExponentialSmoothing(np.asarray(train_df['Count']), trend='add', seasonal='add', seasonal_periods=p)
    sar_model = SARIMAX(np.asarray(train_df['Count']), order=(1, 1, 1), seasonal_order=(1, 1, 1, p))
    exp_model_fit = exp_model.fit()
    sar_model_fit = sar_model.fit()
    exp_forecasts = exp_model_fit.forecast(p)
    sar_forecasts = sar_model_fit.forecast(p)
    avg_forecast = [sar_forecasts[i] + exp_forecasts[i] for i in range(0, p)]
    test_df['Count'].iloc[i*(p): (i+1)*(p)] = avg_forecast
    append_df = test_df.iloc[i*(p): (i+1)*(p),:]
    train_df = pd.concat([train_df, append_df], axis=0)
```
